# Hoja 1

Vamos a dar una vista rapida de los datos

```python
Acc_vel_loc_sinrangos = pd.read_csv("../../data/converted/Acc_vel_loc_sinrangos.csv")
print(Acc_vel_loc_sinrangos.info())
Acc_vel_loc_sinrangos.head(20)
```

```python
Acc_vel_loc_sinrangos.describe()
```

Vamos a rellenar esos datos nulos para que al menos podamos realizar la suma de conexiones totales por cada rango de servicio contratado

```python
Acc_vel_loc_sinrangos
```

Voy a mirar si hay algunas columnas con datos, donde 'Link Indec' aparece como 'sin datos', caso contrario elimino las filas  
- Con describe voy a ver muy por encima si encuentro algun dato anormal

```python
link_indec = Acc_vel_loc_sinrangos[Acc_vel_loc_sinrangos['Link Indec']=='Sin Datos']
link_indec.describe()
```

```python
Acc_vel_loc_sinrangos['Unnamed: 0'].value_counts()
```

La columna 'Unnamed: 0', tiene valores faltantes, son 3065 filas y aparece una longitud de 2995, al considerarla una columna poco util la elimino directamente

```python
Acc_vel_loc_sinrangos = Acc_vel_loc_sinrangos.drop('Unnamed: 0', axis=1)
```

```python
Acc_vel_loc_sinrangos.Localidad.value_counts()
```

Considero que al no conocer la ubicacion de la localidad, no ayudaria al analisis de la ubicacion y el servicio que se requiere en el lugar.  
La voy a eliminar para no tener una agrupacion erronea de las localidades

```python
Acc_vel_loc_sinrangos = Acc_vel_loc_sinrangos[Acc_vel_loc_sinrangos.Localidad!='Otros']
Acc_vel_loc_sinrangos.Localidad.value_counts()
```

cabe mencionar que hay localidades que coinciden con el nombre de otras localidades ubicadas en otras provincias.  
- por ejemplo:

```python
san_jose = Acc_vel_loc_sinrangos[Acc_vel_loc_sinrangos.Localidad=='San José']
san_jose
```

# Hoja 2

```python
Accesos_Por_Tecnología = pd.read_csv('../../data/converted/Accesos Por Tecnología.csv')
Accesos_Por_Tecnología.info()
```

Doy un primer vistazo al DF

```python
Accesos_Por_Tecnología = Accesos_Por_Tecnología.drop('Unnamed: 0', axis=1)
Accesos_Por_Tecnología
```

A primera vista se logra encontrar filas nulas.  
Voy a revisar primero la ubicacion de la info(Provincia)

```python
provincia_desconocida = Accesos_Por_Tecnología[Accesos_Por_Tecnología.Provincia.isna()]
provincia_desconocida
```

Elimino las filas con varios valores nulos

```python
Accesos_Por_Tecnología = Accesos_Por_Tecnología.dropna(subset=['Provincia'])
```

Ahora es el turno de Trimestre

```python
Accesos_Por_Tecnología.Trimestre.value_counts()
```

hay 3 valores mal cargados, vamos a revisarlos

```python
trimestre = Accesos_Por_Tecnología[Accesos_Por_Tecnología.Trimestre=='3 *']
trimestre
```

```python
trimestre = Accesos_Por_Tecnología[Accesos_Por_Tecnología.Trimestre=='2 *']
trimestre
```

```python
trimestre = Accesos_Por_Tecnología[Accesos_Por_Tecnología.Trimestre=='1 *']
trimestre
```

Los 3 trimestres corresponden al año 2019, el cual tambien tiene un asterisco (que vamos a revisar luego), antes de modificarlo voy a revisar el año mismo para chequear si tiene datos faltantes

```python
anio = Accesos_Por_Tecnología[Accesos_Por_Tecnología['Año']=='2019']
```

```python
anio
```

Habiendo chequeado que faltan filas voy a corregir el error en los trimestres, reemplazando en el string de trimestre el '1 *' por un '1'

```python
Accesos_Por_Tecnología.Trimestre = Accesos_Por_Tecnología.Trimestre.str.replace(' *','') 
```

```python
Accesos_Por_Tecnología.Trimestre.value_counts()
```

Finalmente revisamos la columna Año

```python
Accesos_Por_Tecnología.Año.value_counts()
```

Reemplazamos el valor mal cargado

```python
Accesos_Por_Tecnología.Año = Accesos_Por_Tecnología.Año.str.replace(' *','')
```

```python
Accesos_Por_Tecnología.Año.value_counts()
```

```python
Accesos_Por_Tecnología.to_parquet('../../data/interim/accesos_tecnologia')
```
